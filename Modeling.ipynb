```python
import numpy as np 
import pandas as pd 
```

```python
#Load Dataset
df = pd.read_csv('dataset.csv')
df
```

```python
#View number of missing values per column
df.isnull().sum(axis=0)
```

```python
#Fill in missing values
df['Title'] = df['Title'].fillna('')
df['Review Text'] = df['Review Text'].fillna('')
df['Division Name'] = df['Division Name'].fillna('Missing')
df['Department Name'] = df['Department Name'].fillna('Missing')
df['Class Name'] = df['Class Name'].fillna('Missing')
```

```python
df = df.drop(['Unnamed: 0', 'Clothing ID', 'Rating'], axis=1)
```

```python
#Lengths of reviews
def length(s):
    return len(s.split())
    
df['Review Length'] = df['Review Text'].apply(length)
```

```python
#Sentiment polarity/subjectivity
from textblob import TextBlob

def polarity(s):  
    return TextBlob(s).sentiment.polarity   
def subjectivity(s):
    return TextBlob(s).sentiment.subjectivity

df['review polarity'] = df['Review Text'].apply(polarity)
df['title polarity'] = df['Title'].apply(polarity)
df['review subjectivity'] = df['Review Text'].apply(subjectivity)
df['title subjectivity'] = df['Title'].apply(subjectivity)
```

```python
df.columns
```

```python
#Split into train and test set
from sklearn.model_selection import train_test_split

X = df.drop('Recommended IND', axis=1)
y = df['Recommended IND']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
```

```python
X_train['Division Name'].unique()
```

```python
#Create dummy variables for Division Name, Department Name, and Class Name columns
for division in X_train['Division Name'].unique():
    X_train['Division Name is ' + division] = X_train['Division Name'].apply(lambda k: 1 if k == division else 0)
    X_test['Division Name is ' + division] = X_test['Division Name'].apply(lambda k: 1 if k == division else 0)
for department in X_train['Department Name'].unique():
    X_train['Department Name is ' + department] = X_train['Department Name'].apply(lambda k: 1 if k == department else 0)
    X_test['Department Name is ' + department] = X_test['Department Name'].apply(lambda k: 1 if k == department else 0)
for c in X_train['Class Name'].unique():
    X_train['Class Name is ' + c] = X_train['Class Name'].apply(lambda k: 1 if k == c else 0)
    X_test['Class Name is ' + c] = X_test['Class Name'].apply(lambda k: 1 if k == c else 0)
    
```

```python
X_train.columns
```

```python
#Remove original Division Name, Department Name, and Class Name columns
for col in ['Division Name','Department Name','Class Name']:
    X_train = X_train.drop(col, axis=1)
    X_test = X_test.drop(col, axis=1)
```

```python
#Normalize columns
columns = ['Age','Positive Feedback Count','Review Length','review polarity',
          'title polarity','review subjectivity','title subjectivity']
for c in columns:
    m = np.mean(X_train[c])
    sd = np.std(X_train[c])
    X_train[c] = (X_train[c] - m) / sd
    X_test[c] = (X_test[c] - m) / sd
```

```python
#Isolate review text column
Review_train, Review_test = X_train['Review Text'], X_test['Review Text']
```

```python
#Preprocess the review text
from nltk.corpus import stopwords 

stop_words = set(stopwords.words('english'))
def preprocessing(text):
    #Remove Stopwords
    newstring = ' '.join([word for word in text.split() if word not in stop_words]) 
    return newstring

Review_train = Review_train.apply(preprocessing)
Review_test = Review_test.apply(preprocessing)
```

```python
Review_train.iloc[0]
```

```python
#Create count vectors from documents
from sklearn.feature_extraction.text import CountVectorizer

vectorizer_1a = CountVectorizer()
vectorizer_1a.fit(Review_train)
Review_train_a = vectorizer_1a.transform(Review_train)
Review_test_a = vectorizer_1a.transform(Review_test)
```

```python
Review_train_a.shape
```

```python
#Create tfidf vectors from documents
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer_1b = TfidfVectorizer()
vectorizer_1b.fit(Review_train)
Review_train_b = vectorizer_1b.transform(Review_train)
Review_test_b = vectorizer_1b.transform(Review_test)
```

```python
#Fit initial model
from sklearn.linear_model import LogisticRegression

model_a = LogisticRegression(solver='lbfgs', max_iter=5000)
model_a.fit(Review_train_a, y_train)
print('model a')
print(model_a.score(Review_test_a, y_test))
print('')
model_b = LogisticRegression(solver='lbfgs', max_iter=5000)
model_b.fit(Review_train_b, y_train)
print('model b')
print(model_b.score(Review_test_b, y_test))
```

```python
#Fit some more models
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.linear_model import SGDClassifier
from sklearn.naive_bayes import ComplementNB
from sklearn.ensemble import RandomForestClassifier

def report(X_train, X_test, y_train, y_test, naive_bayes=True):
    model = LogisticRegression(solver='lbfgs', max_iter=5000)
    model.fit(X_train, y_train)
    print('Logistic Regression: {}'.format(str(model.score(X_test, y_test))))

    model = SVC(gamma='scale')
    model.fit(X_train, y_train)
    print('SVC: {}'.format(str(model.score(X_test, y_test))))

    model = SGDClassifier(max_iter=1000, tol=-np.inf)
    model.fit(X_train, y_train)
    print('SGDClassifier: {}'.format(str(model.score(X_test, y_test))))

    from sklearn.neighbors import KNeighborsClassifier
    model = KNeighborsClassifier()
    model.fit(X_train, y_train)
    print('KNeighborsClassifier: {}'.format(str(model.score(X_test, y_test))))
    
    if naive_bayes:
        model = ComplementNB()
        model.fit(X_train, y_train)
        print('Naive Bayes: {}'.format(str(model.score(X_test, y_test))))

    model = RandomForestClassifier(n_estimators=100)
    model.fit(X_train, y_train)
    print('Random Forest: {}'.format(str(model.score(X_test, y_test))))

print('Count vectors')
report(Review_train_a, Review_test_a, y_train, y_test)
print()
print('tfidf vectors')
report(Review_train_b, Review_test_b, y_train, y_test)
```

```python
#Model for review text
model_1 = LogisticRegression(solver='lbfgs', max_iter=5000)
model_1.fit(Review_train_a, y_train)
print('Logistic Regression: {}'.format(str(model_1.score(Review_test_a, y_test))))
```

```python
#Preprocess the title text
T_train, T_test = X_train['Title'], X_test['Title']

stop_words = set(stopwords.words('english'))
def preprocessing(text):
    #Remove Stopwords
    newstring = ' '.join([word for word in text.split() if word not in stop_words]) 
    return newstring

T_train = T_train.apply(preprocessing)
T_test = T_test.apply(preprocessing)
```

```python
#Count vectors and tfidf vectors for titles 
vectorizer_2a = CountVectorizer()
vectorizer_2a.fit(T_train)
T_train_a = vectorizer_2a.transform(T_train)
T_test_a = vectorizer_2a.transform(T_test)

vectorizer_2b = TfidfVectorizer()
vectorizer_2b.fit(T_train)
T_train_b = vectorizer_2b.transform(T_train)
T_test_b = vectorizer_2b.transform(T_test)
```

```python
#Fit some models using titles as feature
print('Count vectors')
report(T_train_a, T_test_a, y_train, y_test)
print()
print('tfidf vectors')
report(T_train_b, T_test_b, y_train, y_test)
```

```python
#model for title text
model_2 = LogisticRegression(solver='lbfgs', max_iter=5000)
model_2.fit(T_train_a, y_train)
print('Logistic Regression: {}'.format(str(model_2.score(T_test_a, y_test))))
```
